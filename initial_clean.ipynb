```python
import pandas as pd
```

```python
#import raw data
df=pd.read_csv("Data/Sleep_health_and_lifestyle_dataset.csv")
df.head()
```

```python
#Drop the person_id column
df=df.drop(columns = "Person ID")
df.head()
```

```python
# Get all data types
df.dtypes
```

```python
# Split 'blood_pressure' on '/' into two seperate columns and convert to integers.
df[['systolic blood pressure', 'diastolic blood pressure']] = df['Blood Pressure'].str.split('/', n=1, expand= True)
df = df.drop(columns = 'Blood Pressure')
df = df.astype({'systolic blood pressure': int, 'diastolic blood pressure': int})
df.head()
```

```python
#set personID as index (now depreciated)
#df=df.set_index("Person ID")
#df.head()
```

```python
#Pandas was sometimes electing to replace "None" in the Sleep Disorder column with NaN
#This ensures the data is preserved properly
df["Sleep Disorder"].fillna("None", inplace=True)
df.head()
```

```python
df.to_csv("Data/Sleep_health_mk1.csv")
```
